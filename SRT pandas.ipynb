```python
x = srt.pivot_table(index = None, columns = 'Competition',values = ['Test','ODI','FC','LA'],aggfunc='first')
```

```python
srt1 = pd.DataFrame(x)
```

```python
srt1.dtypes
```

```python
srt1['Runs scored'] = srt1['Runs scored'].astype('int')
srt1 = srt1.astype({"10 wickets in match": int, "Balls bowled": int, "Batting average": float, "Bowling average": float, "Matches": int, "Wickets": int})
```
